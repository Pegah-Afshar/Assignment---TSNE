```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
from sklearn.datasets import fetch_openml
```

1. Load the dataset and conduct any necessary preprocessing, such as normalizing the data.

```python
mnist = fetch_openml('Fashion-MNIST', version=1, cache=True)
```

```python
indice=np.random.choice(70000,10000)
X=mnist.data[indice]/255
y=mnist.target[indice]
print(X.shape,y.shape)
```

```python
plt.figure(figsize=(20,5))

plt.subplot(141)
plt.imshow(X[123].reshape(28,28))
plt.title("Label of the image: {}".format(y[123]))

plt.subplot(142)
plt.imshow(X[124].reshape(28,28))
plt.title("Label of the image: {}".format(y[124]))

plt.subplot(143)
plt.imshow(X[125].reshape(28,28))
plt.title("Label of the image: {}".format(y[125]))

plt.subplot(144)
plt.imshow(X[126].reshape(28,28))
plt.title("Label of the image: {}".format(y[126]))


plt.show()
```

2. Apply t-SNE to the data by setting n_components=2

```python
from sklearn.manifold import TSNE
tsne=TSNE(n_components=2,n_iter=300)
tsne_result=tsne.fit_transform(X)
```

3. Using the two-dimensional t-SNE representation, draw a graph of the data by coloring and labeling the data points as you did in the checkpoint.

```python

plt.figure(figsize=(10,5))
colours = ["r","b","g","c","m","y","k","r","burlywood","chartreuse"]
for i in range(tsne_result.shape[0]):
    plt.text(tsne_result[i, 0], tsne_result[i, 1], str(y[i]),
             color=colours[int(y[i])],
             fontdict={'weight': 'bold', 'size': 50}
        )

plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(tsne_result[:, 0], tsne_result[:, 1])
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```
